## Stabiliser operator

````python
import numpy as np
import sympy as smp


zero_one = np.array([[1],[0]])
I = np.array([[1,0], [0,1]])
Z = np.array([[1, 0],[0, -1]])
X = np.array([[0,1],[1,0]])
H = np.array([[1, 1],[1, -1]]*1/np.sqrt(2))
Y = np.array([[0,-1j],[1j,0]])

three_qbt = smp.Matrix((np.kron(zero_one, np.kron(zero_one, zero_one))))
list_of_arrays = [I, Z, X, H, Y]
list_length = len(list_of_arrays)
dict1 = {0:'I', 1:'Z', 2:'X', 3:'H', 4:'Y'}
for i in range(list_length):
    for j in range(list_length):
        for k in range(list_length):
            if smp.Matrix(np.kron(list_of_arrays[k], np.kron(list_of_arrays[i], list_of_arrays[j])))*three_qbt == three_qbt:
                lst = [dict1[matrix] for matrix in [i, j, k]]
                print(''.join(lst) + ' is a stabilizer operator')
````

# Chapter 3.6 Phase estimation circuit

````python
#initialization
import matplotlib.pyplot as plt
import numpy as np
import math

# importing Qiskit
from qiskit import IBMQ, Aer, transpile, assemble
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister

# import basic plot tools
from qiskit.visualization import plot_histogram

# Count is selected based on the desired error rate
count = 3

qpe = QuantumCircuit(count + 1, count)


# Eigenstate Initialisation
qpe.x(count)

# Counting Qubit Initilisation
for qubit in range(count):
    qpe.h(qubit)
    
    
    
# Unitary gate initilisation, apply 2 x of desired theta
theta = math.pi*(2/3)
repetitions = 1
for counting_qubit in range(count):
    for i in range(repetitions):
        qpe.cp(theta, counting_qubit, count); # This is CU
    repetitions *= 2

# Inverse of QFT
def qft_dagger(qc, n):
    """n-qubit QFTdagger the first n qubits in circ"""
    # Don't forget the Swaps!
    for qubit in range(n//2):
        qc.swap(qubit, n-qubit-1)
    for j in range(n):
        for m in range(j):
            qc.cp(-math.pi/float(2**(j-m)), m, j)
        qc.h(j)
        

qpe.barrier()
# Apply inverse QFT
qft_dagger(qpe, count)
# Measure
qpe.barrier()
for n in range(count):
    qpe.measure(n,n)

qpe.draw()
qpe.decompose().draw()
````

````python
aer_sim = Aer.get_backend('aer_simulator')
shots = 2048
t_qpe = transpile(qpe, aer_sim)
qobj = assemble(t_qpe, shots=shots)
results = aer_sim.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer,figsize=(20,11))
````

````python
from qiskit.providers.aer.noise import NoiseModel
from qiskit.providers.aer.noise.errors import pauli_error, depolarizing_error
from qiskit import QuantumRegister, ClassicalRegister, execute
from qiskit import QuantumCircuit, Aer, transpile, assemble
from qiskit.visualization import plot_histogram
aer_sim = Aer.get_backend('aer_simulator')

def get_noise(p_meas,p_gate):
    # p_meas prob to switch 0 and 1 for measurement
    # P_gate prob to replace qubit with random state
    error_meas = pauli_error([('X',p_meas), ('I', 1 - p_meas)])
    error_gate1 = depolarizing_error(p_gate, 1)
    error_gate2 = error_gate1.tensor(error_gate1)

    noise_model = NoiseModel()
    noise_model.add_all_qubit_quantum_error(error_meas, "measure") # measurement error is applied to measurements
    noise_model.add_all_qubit_quantum_error(error_gate1, ["x"]) # single qubit gate error is applied to x gates
    noise_model.add_all_qubit_quantum_error(error_gate2, ["cx"]) # two qubit gate error is applied to cx gates
        
    return noise_model

noise_model = get_noise(0.01,0.01)
display(qc0.decompose().draw())

qc0 = QuantumCircuit(3) # initialize circuit with three qubits in the 0 state
qc0.measure_all() # measure the qubits

# run the circuit with the noise model and extract the counts

qobj = assemble(qc0)
counts = aer_sim.run(qobj, noise_model=noise_model).result().get_counts()

# Iterate
print(counts)

# Multiple tries
for i in range(10):
    counts_dict = aer_sim.run(qobj, noise_model=noise_model).result().get_counts()
    print(counts_dict)
    for j in counts_dict.keys():
        if j in counts.keys():
            counts[j] += counts_dict[j]
        else:
            counts[j] = counts_dict[j]


# change Shot size (cant do??)
#job_sim = execute([qc0, qc0], aer_sim, shots=2e6) #usually 1e6 max shot size

#counts = job_sim.result().get_counts()


plot_histogram(counts)
````

````python
#initialization
import matplotlib.pyplot as plt
import numpy as np
import math

# importing Qiskit
from qiskit import IBMQ, Aer, transpile, assemble
from qiskit import QuantumCircuit, ClassicalRegister, QuantumRegister

# import basic plot tools
from qiskit.visualization import plot_histogram

# Count is selected based on the desired error rate
count = 4

qpe = QuantumCircuit(count + 3, count)


# Eigenstate Initialisation
qpe.h(count)

for i in range(count):
    qpe.h(i)
# Unitary gate initilisation, apply 2 x of desired theta
#theta = math.pi*(2/3)
repetitions = 1
for counting_qubit in range(count):
    for i in range(repetitions):
        qpe.cz(counting_qubit, 4)
        qpe.cz(counting_qubit, 5)
    repetitions *= 2

# Inverse of QFT
def qft_dagger(qc, n):
    """n-qubit QFTdagger the first n qubits in circ"""
    # Don't forget the Swaps!
    for qubit in range(n//2):
        qc.swap(qubit, n-qubit-1)
    for j in range(n):
        for m in range(j):
            qc.cp(-math.pi/float(2**(j-m)), m, j)
        qc.h(j)
        

qpe.barrier()
# Apply inverse QFT
qft_dagger(qpe, count)
# Measure
qpe.barrier()
for n in range(count):
    qpe.measure(n,n)

qpe.draw()
#qpe.decompose().draw()
````

````python
aer_sim = Aer.get_backend('aer_simulator')
shots = 2048
t_qpe = transpile(qpe, aer_sim)
qobj = assemble(t_qpe, shots=shots)
results = aer_sim.run(qobj).result()
answer = results.get_counts()

plot_histogram(answer,figsize=(20,11))
````
