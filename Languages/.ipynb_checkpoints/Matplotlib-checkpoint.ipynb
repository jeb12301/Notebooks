```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.rcParams['figure.dpi'] = 600
```

```python
np.exp(1)
```

```python
t = np.linspace(0.01, 2.0, num=1000)
y = np.exp(1/t)/t**2*(1/(np.exp(1/t)-1))**2
```

```python
plt.plot(t, y)
plt.title('$C/Nk$ against parameter $t$ from 0 to 2 seconds')
plt.ylabel('$C/Nk$')
plt.xlabel('$t$')
plt.savefig('plot.png', dpi=400)
```
