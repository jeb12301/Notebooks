# Generating Evolution under Tk (Qubit)

```python
def modadd(state, Z_or_X, index):
    if state[Z_or_X][index]:
        state[Z_or_X][index]=0
    else:
        state[Z_or_X][index]=1
    return state

def evolution(state, size):
    Z_part, X_part = state
    new_state = [[0 for _ in range(size)],[0 for _ in range(size)]]
    # X to Z evolution
    for i in range(size):
        if X_part[i]:
            # Z
            new_state = modadd(new_state, 0, i)
    # Z evolution
    for i in range(size):
        if Z_part[i]:
            if i == 0: # Boundary
                # Z
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i+1)
                # X
                new_state = modadd(new_state, 1, i)
            elif i < size-1:
                # Z
                new_state = modadd(new_state, 0, i-1)
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i+1)
                # X
                new_state = modadd(new_state, 1, i)
            else:
                # Z
                new_state = modadd(new_state, 0, i-1)
                new_state = modadd(new_state, 0, i)
                # X
                new_state = modadd(new_state, 1, i) 
    return new_state
```

```python
state_0=[[1, 0, 0, 0],[0, 0, 0, 0]] # (Z, X)

size = len(state_0[0])

state = evolution(state_0, size)

all_states = [state_0]

while state != state_0:
    all_states.append(state)
    state = evolution(state, size)

print(len(all_states))
```

```python
def modadd(state, Z_or_X, index):
    if state[Z_or_X][index] == dim-1:
        state[Z_or_X][index]=0
    else:
        state[Z_or_X][index]+=1
    return state

def evolution(state, size):
    Z_part, X_part = state
    new_state = [[0 for _ in range(size)],[0 for _ in range(size)]]
    # X to Z evolution
    for i in range(size):
        for k in range(X_part[i]):
            # Z
            new_state = modadd(new_state, 0, i)
            new_state = modadd(new_state, 0, i)
    # Z evolution
    for i in range(size):
        for k in range(Z_part[i]):
            if i == 0: # Boundary
                # Z
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i+1)
                new_state = modadd(new_state, 0, i+1)
                # X
                new_state = modadd(new_state, 1, i)
            elif i < size-1:
                # Z
                new_state = modadd(new_state, 0, i-1)
                new_state = modadd(new_state, 0, i-1)
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i+1)
                new_state = modadd(new_state, 0, i+1)
                # X
                new_state = modadd(new_state, 1, i)
            else:
                # Z
                new_state = modadd(new_state, 0, i-1)
                new_state = modadd(new_state, 0, i-1)
                new_state = modadd(new_state, 0, i)
                new_state = modadd(new_state, 0, i)
                # X
                new_state = modadd(new_state, 1, i) 
    return new_state
```

```python
def normalise(mat, dim):
    return mat/(np.sqrt(np.trace(mat.transpose().conjugate()*mat)/dim))
```

```python
k = size # Qubit Count
dimension = 2 # Qubit
d = dimension**k

ortho = []
new_ortho = ['start']
commutations = []

for p_mat in p_set:
    new = p_mat.copy()
    count = 0
    for i in ortho:
        if np.array_equal(p_mat, i) or np.array_equal(-p_mat, i):
            count = 1
            break
    if new.any():
        ortho.append(normalise(new, d))
more = True

while more:
    if new_ortho==['start']:
        for mat1 in ortho:
            for mat2 in ortho:
                conj = mat1*mat2-mat2*mat1
                if conj.any():
                    commutations.append(normalise(conj, d))
    else:
        comb1 = list(combinations([i for i in range(len(new_ortho))], 2))
        list1 = [i for i in range(len(new_ortho))]
        list2 = [i for i in range(len(new_ortho), len(ortho))]
        comb2 = list(product(list1,list2))
        total_comb = comb1 + comb2
        for i1, i2 in total_comb:
                conj = ortho[i1]*ortho[i2]-ortho[i2]*ortho[i1]
                if conj.any():
                    commutations.append(normalise(conj, d))     
    new_ortho = []
    for mat_c in commutations:
        if mat_c.any():
            count = 0
            for i in ortho:
                if np.array_equal(mat_c, i) or np.array_equal(-mat_c, i):
                    count = 1
                    break
            for i in new_ortho:
                if np.array_equal(mat_c, i) or np.array_equal(-mat_c, i):
                    count = 1
                    break
            if count == 0:
                new_ortho.append(mat_c)
                print(len(ortho))
    commutations = []
    if len(new_ortho) == 0:
        more = False
    ortho = new_ortho.copy() + ortho.copy()
```

### Generating Arbitrary Gell Mann Matrices

Based on https://mathworld.wolfram.com/GeneralizedGell-MannMatrix.html

**What does it mean?**

You want to find basis for hermitian matrices, so need to populate non-diagonal entries with real value (same sign) and complex numbers (different sign)

Only need n-1 real diagonal matrices to span the n length diagonal for real values

```python
n = 6 # Dimension

Gell_mann1 = []

for k in range(1, n):
    for j in range(k):
        zero_mat = np.zeros((n, n))
        Ekj = np.matrix(zero_mat.copy())
        Ejk = np.matrix(zero_mat.copy())
        Ekj[k,j] = 1
        Ejk[j,k] = 1
        Gell_mann1.append(np.matrix(Ekj +Ejk))


Gell_mann2 = []
for k in range(1, n):
    for j in range(k):
        zero_mat = np.zeros((n, n))
        Ekj = np.matrix(zero_mat.copy())
        Ejk = np.matrix(zero_mat.copy())
        Ekj[k,j] = 1
        Ejk[j,k] = 1
        Gell_mann2.append(np.matrix(Ekj - Ejk))

Gell_mann3 = []
for l in range(1, n):
    const = np.sqrt(2/(l*(l+1)))
    final_mat = np.zeros((n, n))
    for j in range(1, l+1):
        zero_mat = np.zeros((n, n))
        Ejj = np.matrix(zero_mat.copy())
        Ell = np.matrix(zero_mat.copy())
        Ejj[j-1,j-1] = 1
        Ell[l,l] = 1
        final_mat += Ejj-Ell # Note that most texts wrong
    Gell_mann3.append(np.matrix(const*final_mat))
```

```python
p_set = Gell_mann1 + Gell_mann2 + Gell_mann3
```
